```python
import tensorflow as tf
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt

# --- Step 1: Load and Prepare the CIFAR-10 Dataset ---
(x_train, y_train), (_, _) = tf.keras.datasets.cifar10.load_data()
x_train = x_train / 255.0  # Normalize to [0, 1]
# Filter for cats (label 3) and dogs (label 5)
cat_dog_indices = np.where(np.isin(y_train, [3, 5]))[0]
x_train = x_train[cat_dog_indices]
y_train = y_train[cat_dog_indices]

# Create a TensorFlow Dataset
train_dataset = tf.data.Dataset.from_tensor_slices(x_train).shuffle(10000).batch(64)

# --- Step 2: Updated Generator with Convolutional Layers ---
def build_generator():
    inputs = layers.Input(shape=(100,))
    x = layers.Dense(8 * 8 * 256, use_bias=False)(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU()(x)

    x = layers.Reshape((8, 8, 256))(x)

    x = layers.Conv2DTranspose(128, (5, 5), strides=(2, 2), padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU()(x)

    x = layers.Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU()(x)

    outputs = layers.Conv2DTranspose(3, (5, 5), strides=(1, 1), padding='same', use_bias=False, activation='tanh')(x)

    return tf.keras.Model(inputs, outputs)

# --- Step 3: Define the Discriminator 
def build_discriminator():
    input_image = layers.Input(shape=(32, 32, 3))

    x = layers.Conv2D(64, (3, 3), strides=(2, 2), padding='same')(input_image)
    x = layers.LeakyReLU(negative_slope=0.2)(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Conv2D(128, (3, 3), strides=(2, 2), padding='same')(x)
    x = layers.LeakyReLU(negative_slope=0.2)(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Flatten()(x)
    x = layers.Dense(512)(x)
    x = layers.LeakyReLU(negative_slope=0.2)(x)

    output = layers.Dense(1, activation='sigmoid')(x)

    return tf.keras.Model(input_image, output)

# --- Step 4: Define the Loss Functions and Optimizers ---
cross_entropy = tf.keras.losses.BinaryCrossentropy()

generator = build_generator()
discriminator = build_discriminator()

# Lower learning rate for stability
generator_optimizer = tf.keras.optimizers.Adam(1e-5)
discriminator_optimizer = tf.keras.optimizers.Adam(1e-5)

# --- Step 5: Training Functions ---
@tf.function
def train_step(real_images):
    noise = tf.random.normal([tf.shape(real_images)[0], 100])  # Generate noise with the same batch size as real_images
    generated_images = generator(noise, training=True)

    # Add label smoothing: 0.9 for real, slightly higher than 0 for fake
    real_labels = tf.random.uniform(shape=[tf.shape(real_images)[0],], minval=0.9, maxval=1.0)
    fake_labels = tf.random.uniform(shape=[tf.shape(generated_images)[0],], minval=0.0, maxval=0.1)

    real_output = discriminator(real_images, training=True)
    fake_output = discriminator(generated_images, training=True)

    # Calculate losses
    gen_loss = cross_entropy(tf.ones_like(fake_output), fake_output)  # Loss for generator
    disc_loss = cross_entropy(real_labels, real_output) + cross_entropy(fake_labels, fake_output)  # Loss for discriminator

    # Calculate gradients and update weights
    with tf.GradientTape() as tape_gen:
        fake_output = discriminator(generator(noise, training=True), training=True)
        gen_loss = cross_entropy(tf.ones_like(fake_output), fake_output)

    gradients_of_generator = tape_gen.gradient(gen_loss, generator.trainable_variables)
    generator_optimizer.apply_gradients(zip(gradients_of_generator, generator.trainable_variables))

    with tf.GradientTape() as tape_disc:
        real_output = discriminator(real_images, training=True)
        fake_output = discriminator(generated_images, training=True)

        disc_loss = cross_entropy(real_labels, real_output) + cross_entropy(fake_labels, fake_output)

    gradients_of_discriminator = tape_disc.gradient(disc_loss, discriminator.trainable_variables)
    discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))

    return gen_loss, disc_loss

# --- Step 6: Train the Model ---
EPOCHS = 150
seed = tf.random.normal([16, 100])  # For generating images
generator_losses = []
discriminator_losses = []

def generate_and_save_images(model, epoch, test_input):
    predictions = model(test_input, training=False)
    predictions = (predictions + 1) / 2  # Rescale to [0, 1]

    plt.figure(figsize=(4, 4))
    for i in range(predictions.shape[0]):
        plt.subplot(4, 4, i + 1)
        plt.imshow(predictions[i, :, :, :])
        plt.axis('off')
    plt.savefig(f'gan_generated_epoch_{epoch}.png')
    plt.show()

# --- Step 7: Start Training ---
for epoch in range(EPOCHS):
    for image_batch in train_dataset:
        g_loss, d_loss = train_step(image_batch)
        generator_losses.append(g_loss)
        discriminator_losses.append(d_loss)

    # Print the average losses for the epoch
    avg_g_loss = tf.reduce_mean(generator_losses[-len(train_dataset):])
    avg_d_loss = tf.reduce_mean(discriminator_losses[-len(train_dataset):])

    print(f'Epoch {epoch + 1}, Generator Loss: {avg_g_loss:.4f}, Discriminator Loss: {avg_d_loss:.4f}')

    # Generate and save images every 10 epochs, starting from epoch 0
    if (epoch + 1) % 10 == 0 or epoch == 0:
        print(f"Generating and saving images for epoch {epoch + 1}")
        generate_and_save_images(generator, epoch + 1, seed)

# --- Step 8: Plot the Losses ---
plt.figure(figsize=(10, 5))
plt.plot(generator_losses, label='Generator Loss')
plt.plot(discriminator_losses, label='Discriminator Loss')
plt.title("Generator and Discriminator Loss During Training")
plt.xlabel("Training Steps")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
